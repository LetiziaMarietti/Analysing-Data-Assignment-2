## 1. Which variable is the dependent, which the independent variable?

Since we are trying to assess whether any of the literary measures are a good predictor for the success of the story, *kudos* is the dependent variable (y) - the one we measure.  Therefore, *Automated Readability Index (ARI), Flesch Grade (FG), Flesch Ease (FE), Mean Segmental Type-Token Ratio (MSTTR), Word Count, Comment Count* are the independent ones (x), or predictors.

## 2. Use matplotlib to plot scatterplots of the data.

```python
# Differentiate between dependent variable and multiple independent variables
independent_variables = ['ari', 'fg', 'fe', 'msttr', 'words', 'comm']
dependent_variable = 'num' # Kudos

for var in independent_variables:
    plt.scatter(df[var], df['num'], alpha=0.5)
    plt.title(f"Scatterplot of {var} vs Kudos")
    plt.xlabel(var)
    plt.ylabel("num")
    plt.grid(True)
    plt.show()
```

## 3.Perform Simple Linear Regression on the data in the columns
## 4. For each independent variable, plot a function based on the intercept and slope of the linear function.
## 5. Calculate the r2 score using the predefined sklearn.metrics function. 

```python
for ind_var in independent_variables:
    # Define predictor and response
    X = df[[ind_var]] # 2D array (DataFrame)
    Y = df['num'] # 1D array (Series)

    # Create and fit the linear regression model
    lr = LinearRegression()
    regression_model = lr.fit(X, Y)
    
    # use the model to predict kudos from the independent variable 
    y_pred = regression_model.predict(X)
    
    # calulate slope and intercept values
    slope = regression_model.coef_[0]
    intercept = regression_model.intercept_

    # Create a range of values for X to plot the regression line
    x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)  # Generates 100 points between min and max of X
    y_range = slope * x_range + intercept  # Calculate the corresponding Y values using the regression line equation
    
    # calculate R2 score 
    r2 = r2_score(Y, y_pred)

    # Print intercept, slope, and R2 values
    print(f"Variable: {ind_var}")
    print(f"Intercept: {intercept}")
    print(f"Slope: {slope}")
    print(f"R2: {r2}")

    # Plot the data points
    plt.scatter(X, Y, alpha=0.5)

    # Plot the regression line
    plt.plot(x_range, y_range, color= 'red',label='Regression Line') 
    plt.title(f"Scatterplot of {ind_var} vs num with Regression Line")
    plt.xlabel(ind_var)
    plt.ylabel("num")
    plt.legend()
    plt.show()

    print("\n")
```

## 6. For each dependent variable, plot the residuals, too. What can you determine from those plots?
## 7. On the values of the residuals, perform the Shapiro-Wilk test.

```python
for ind_var in independent_variables:

    X = df[[ind_var]]
    Y = df['num']
    linreg = LinearRegression()
    linreg.fit(X, Y)
    y_pred = linreg.predict(X)

    # Calculate the residuals (difference between the observed values of kudos and the predicted ones from the regression model)
    residuals = Y - y_pred

    # Perform Shapiro-Wilk test on the residuals
    shapiro_test = stats.shapiro(residuals)
    print(f"Shapiro-Wilk Test for {ind_var}:", shapiro_test)

    if shapiro_test.pvalue > 0.05:
        print(f"Residuals for {ind_var} are likely normally distributed.")
    else:
        print(f"Residuals for {ind_var} are not normally distributed.")

    # Plot the residuals as a histogram
    sns.histplot(residuals)
    plt.title(f"Histogram of Linear Regression Residuals for {ind_var}")
    plt.xlabel('Residuals')
    plt.ylabel("Count")
    plt.show()
```
